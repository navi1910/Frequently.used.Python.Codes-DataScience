# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Matplotlib Basics

```python
x = np.random.randn(100,1)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
x = np.array([12,15,17,19])
y = x**2
plt.scatter(x, y, color = 'r', s = x)
plt.title('x vs y')
plt.xlabel('x values')
plt.ylabel('y values')
plt.yticks([100,200,300,400], ['1H', '2H', '3H','4H'])
plt.show()
```

```python
plt.hist(x)
plt.show()
```

# Subplot methods

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")
plt.xlabel("x")
plt.ylabel("y")
plt.title("x vs y")

plt.subplot(1,2,2)
plt.plot(y,x,"g")
plt.xlabel("y")
plt.ylabel("x")
plt.title("y vs x")
plt.show()
```

# Object Oriented Methods

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("X vs Y")
```

```python
fig1 = plt.figure()
axes1 = fig1.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig1.add_axes([0.2, 0.4, 0.3, 0.4])


axes1.plot(x,y)
axes2.plot(y,x, "r")
```

```python
#Object oriented method 02
fig,axes = plt.subplots(2,1, figsize=(8, 5), dpi =100)
axes[0].plot(x,y)
plt.show()
plt.tight_layout()
#fig.savefig("my_picture.png", dpi = 200)
```

```python
fig3 = plt.figure()
ax1 = fig3.add_axes([0.1, 0.1, 0.8, 0.8])
ax1.plot(x, y, label = "X vs Y", linewidth= 2, color = "green", marker = "o", markersize = 10, markerfacecolor = "yellow",
         markeredgecolor = "blue")
ax1.plot(y,x, label = "Y vs X", color = "red", alpha = 0.5, linestyle = "dotted")
plt.legend(loc = 10)

plt.show()
# linewideth = lw
# linestyle = ls 
# loc = 0 is 'best'
# loc = (0.1, 0.1)
```

```python
fig3 = plt.figure()
ax1 = fig3.add_axes([0.1, 0.1, 0.8, 0.8])
ax1.plot(x, y, label = "X vs Y")
ax1.plot(y,x, label = "Y vs X")
ax1.set_ylim([1000,10000])
plt.legend(loc = 10)

plt.show()
```

# Frequency Plots/Distribution Plots
Usually take only one variable

```python
t = sns.load_dataset("tips")
t.head(0)
```

Distplot (Histogram with KDE)

```python
sns.distplot(t["total_bill"], kde = True, bins = 10)
plt.show()
```

KDE plot (Kernel Density Plot)

```python
sns.kdeplot(t["total_bill"])
```

Rugplot
Used to produce KDE

```python
sns.rugplot(t["total_bill"])
```

Joint Plot.
Used to compare two distributions.
Can be of various kinds, hexbins, scatter etc.



```python
sns.jointplot(x = "total_bill", y = "tip", data = t, kind = "reg")
```
